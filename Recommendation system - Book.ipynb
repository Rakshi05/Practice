```python
data = df_clean.rename(columns={"Book.Title":"Book_title"},inplace=False)
data.head()
```

```python
#Convert text into matrix

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import CountVectorizer

cm = CountVectorizer().fit_transform(data["Book_title"])
```

```python
data["Book_title"].shape
```

```python
cm.data.shape
```

```python
cs = cosine_similarity(cm)
```

```python
cs.shape
```

```python
cs
```

```python
new_book = "Decision in Normandy"

book_id = data[data.Book_title==new_book].index.values[0]
```

```python
data[data["Book_title"]=="Decision in Normandy"]
```

```python
book_id
```

```python
cs[2]
```

```python
score = list(enumerate(cs[book_id]))
```

```python
score
```

```python
for i in score:
    print(i[1])
```

```python
sorted_scores = sorted(score, key= lambda x:x[1], reverse= True)
```

```python
sorted_scores
```

```python
sorted_scores[1:]
```

```python
j = 0
print("The 3 most recommended Books to ",new_book,"are:\n")

for item in sorted_scores[1:]:
    Book_title = data[data.index ==item[0]]['Book_title'].values[0]
    print(j+1,Book_title)
    j = j+1
    if j>2:
        break
        
```
